### What is its structure?

The breast cancer dataset consists of 569 rows and 32 columns. Each row is an instance of cell nucleus characteristics, with 30 input linearly separable features and one output variable, i.e., diagnosis.

The features of the dataset are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe the characteristics of the cell nuclei present in the image.

The columns are aggregated into the following descriptions as follows:
1. Radius - These features represent the average distance from the center to points on the perimeter of the cell nuclei in the image.
2. Texture-  These features describe the variation in the gray-scale values of the pixels in the image, which gives an indication of the texture or coarseness of the cell nuclei.
3. Perimeter - These features represent the total perimeter of the cell nuclei in the image.
4. Area - These features represent the total area of the cell nuclei in the image.
5. Smoothness - These features measure the amount of variation in the radius lengths of the cell nuclei. A high value would indicate a rough or irregular surface.
6. Compactness - These features measure the compactness of the cell nuclei, i.e., how closely the nuclei are packed together. A high value would indicate a compact nuclei with little inter-cell space.
7. Concavity - These features describe  the severity of concave portions of the cell nuclei contour. A high value indicates a more concave or irregular shape.
8. Concave points - These features measure the number of concave portions or dents in the cell nuclei contour.
9. Symmetry - These features measure the symmetry of the cell nuclei. A high value indicates a more symmetrical shape.
10. Fractal dimension - These features describe the complexity of the cell nuclei contour. A high value would indicate a more complex or irregular shape.


The field "diagnosis" contains 357 benign (B) and 212 malignant (M) cases. The goal of the project would be to classify the diagnosis as either benign or malignant based on the input features.
